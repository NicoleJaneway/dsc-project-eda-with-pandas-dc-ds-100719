### Box Plot

```python
RiverPlot1 = BostonHousing.boxplot(column='crim',by='chas', grid=False, sym='.');
RiverPlot1.set_xlabel('River Tracts');
RiverPlot1.set_ylabel('Crime');
plt.ylim(0,15)
plt.title('')

RiverPlot1;
```

# Split 2 - Nox level

```python
NoxPlot1 = BostonHousing['nox'].plot(kind='hist',title='Nitrogen Oxide Histogram');
NoxPlot1.set_xlabel('nox value');
NoxPlot1;
```

```python
BostonHousing['nox'].value_counts(bins=5)
```

```python
bins = [.384, .482, .579, .677, .774, .871]
names = ['Low', 'Somewhat low', 'Moderate', 'Somewhat high', 'High']

BostonHousing['nox_cat'] = pd.cut(BostonHousing['nox'], bins, labels = names)
NoxPlot2 = pd.DataFrame(BostonHousing['nox_cat'].value_counts()).loc[names].plot(kind='bar', title='Nitrogen Oxide Histogram');
NoxPlot2.set_xlabel('Nox level');
NoxPlot2.set_ylabel('Frequency');
NoxPlot2;
```

# Split 3 - Age

```python
BostonHousing['age'].value_counts(bins=5)
```

```python
(BostonHousing.loc[BostonHousing['age'] == 100])['age'].count()
```

```python
Post1940 = (BostonHousing.loc[BostonHousing['age'] < 100])
Post1940
```

```python
Post1940['age'].value_counts(bins=4)
```

```python
bins2 = [5,30,50,75,99.3,100]
names2 = ['5-30','30-50','50-75','75-99','100']

BostonHousing['age_cat'] = pd.cut(BostonHousing['age'], bins2, labels=names2)
BostonHousing['age_cat'].value_counts()
```

```python
AgePlot0 = BostonHousing['age'].plot(kind='hist', title='Age Histogram')
```

```python
AgePlot1 = pd.DataFrame(BostonHousing['age_cat'].value_counts()).loc[names2].plot(kind='bar', title='Age Histogram')
```
